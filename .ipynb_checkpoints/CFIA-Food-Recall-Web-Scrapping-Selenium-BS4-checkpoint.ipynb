
# Importing Libraries, Selenium and BeautifulSoup

```python
from selenium import webdriver # Initialize web browser
from selenium.webdriver.support.ui import WebDriverWait # Wait for a page to load

from bs4 import BeautifulSoup

import pandas as pd

#specify driver path
DRIVER_PATH = 'C:\\webdrivers\\chromedriver.exe'
driver = webdriver.Chrome(executable_path = DRIVER_PATH)
```

# Extracting Links and Titles

```python
links = []
recalled = []


for i in range(0,4150,50):
    driver.get('https://healthycanadians.gc.ca/recall-alert-rappel-avis/search-recherche/simple/en?s=&plain_text=&f_mc=1&js_en=&page=50&per_page='+str(i))
    driver.implicitly_wait(2) # Wait 2 seconds to load the page before parsing and extracting the data
    page_source = driver.page_source
    soup = BeautifulSoup(page_source, 'html.parser')
    divs = soup.find_all('div', class_='margin-bottom-medium word_wrap')

    for div in divs:
        item = {
            'title': div.find('a').text.replace('\n\t\t\t\t\t\t\t\t', '').replace('\t\t\t\t\t\t\t', '').strip(),
            'link': 'https://healthycanadians.gc.ca/' + div.find('a').get('href').strip(),
            'date_scrapped': 'February 14, 2021'
        }
        link = 'https://healthycanadians.gc.ca/' + div.find('a').get('href').strip()
        links.append(link)
        recalled.append(item)


recalls = pd.DataFrame(recalled)
recalls
```

```python
# Export and save to .csv file
recalls.to_csv('C:\\Users\jorge\Desktop\cfia_recalls_v2.csv', encoding='utf-16')
```

# Extracting Recall Details

```python
recalls_list = []

for i in recalls['link']:
    driver.get(i)
    driver.implicitly_wait(2)
    page_source = driver.page_source
    soup = BeautifulSoup(page_source, 'html.parser')
    details = soup.find_all('dd', class_ = 'width45 paddingNone')
    name = soup.find(id= 'cn-cont')
    
    try:
        dictionary = {
            'name': name.text.strip(),
            'date': details[0].text.strip(),
            'type': details[1].text.strip(),
            'sub_type': details[2].text.strip(),
            'hazard': details[3].text.strip(),
            'hazard_class': details[4].text.strip(),
            'source_of_recall': details[5].text.strip(),
            'company': details[6].text.strip(),
            'distribution': details[7].text.strip(),
            'channel': details[8].text.strip(),
            'ref_num': details[9].text.strip(),
            'date_scrapped': 'February 15, 2021'
        }
        recalls_list.append(dictionary)
    except:
        pass

recall_details = pd.DataFrame(recalls_list)
recall_details.head()
```

```python
# Export and save to .csv file
recall_details.to_csv('C:\\Users\jorge\Desktop\cfia_recalls_details_v2.csv', encoding='utf-16')
```
